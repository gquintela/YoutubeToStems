# Import everithing needed...

```
from __future__ import unicode_literals
import youtube_dl
import os
import subprocess
from datetime import date
```

# Select your input ...

```
# enter your youtube links here, as strings separated with commas, an awesome example is given!
links = ["https://www.youtube.com/watch?v=edTVgwKZolM"]

# Stems quantity: 2,4 or 5!
stemsQty = '5'
```

# Main code

```
today = date.today()
date = today.strftime("%d-%m-%Y")
folder = "audio_output_" + date

if(stemsQty != '2' and stemsQty !='4' and stemsQty !='5'):
	print("please select correct ammount of stems: 2, 4 or 5.")
	exit(0)

spleeter = "spleeter separate -c mp3 -o " + folder + " -p spleeter:" + stemsQty + "stems "

ydl_opts = {
    'format': 'bestaudio/best',
    'outtmpl': '%(title)s.%(ext)s',
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'mp3',
        'preferredquality': '320',
    }],
}

count = 0
print("\n\n---SONG DOWNLOAD ---\n")

for link in links:
    with youtube_dl.YoutubeDL(ydl_opts) as ydl:
        ydl.download([link])
        print(" ")
    count += 1
 
print(f"\nTotal files downloaded: {count}")
if (not count):
	print(f"\n{args.inputFileName} is empty, please fill it with Youtube links!")
	exit(0)

print("\n\n---TRACK SEPARATION ---\n")

entries = os.listdir('./')
for file in entries:
    newFile = file.replace("(","[").replace(")","]")#.replace(" ","")
    os.rename(file, newFile)
    if ".mp3" in newFile:
        print(f"\nSeparating: {newFile}")
        # print(spleeter + newFile)
        subprocess.call(spleeter + " \"" + newFile + "\"" , shell=True)
        os.remove(newFile)

print(f"\nAll files separated. Stems can be found in {folder}")
```
